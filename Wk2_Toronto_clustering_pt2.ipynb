```python
# Get the of indexes for which Borough is not assigned
dropped_index = df_data_0[ df_data_0['Borough'] == 'Not assigned' ].index
dropped_index
```

```python
# Delete these row indexes from dataFrame
df_data_0.drop(dropped_index , inplace=True)
```

```python
df_data_0.reset_index(drop=True, inplace=True)
df_data_0.head()
```

```python
#Now we want to have a list of each and unique PostalCode
PC = df_data_0['PostalCode'].unique()
PC
```

```python
#To start we will merge all the Neighbourhood with the same Postal Code
#We split this column into a new dataframe df2
df2 = pd.DataFrame(columns=["Neighbourhood"])
```

```python
#We have 104 unique Postal Codes
#iterations = the number of neighbourhood with the same postal code
for i in range (0,103):
    n  = df_data_0.loc[df_data_0['PostalCode'] == PC[i],['Neighbourhood']]
    iterations = df_data_0.loc[df_data_0['PostalCode'] == PC[i],['Neighbourhood']].get("Neighbourhood").count()
    j = 0
    concat =""
    if iterations == 1:
        concat = n.iloc[0,0]
    else:
        concat = n.iloc[0,0]
        while j < iterations-1:
            concat = concat + ", " + n.iloc[j+1,0]
            j = j + 1
    df2 = df2.append({'Neighbourhood': concat}, ignore_index=True)
```

```python
df2.head()
```

```python
#Then we drop the rows with duplicates Borough
df = df_data_0[['PostalCode','Borough']]
df.drop_duplicates(inplace = True)
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
df.tail(10)
```

```python
#We now concatenate df(PostalCode + Borough) and df2(Neighbourhood)
df3 = pd.concat([df, df2], axis=1)
df3.head()
```
